```python
ACCESS_TOKEN = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
ACCESS_SECRET = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
CONSUMER_KEY = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
CONSUMER_SECRET = 'XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX'
```

ACCESS_TOKEN,ACCESS_SECRET,CONSUMER_KEY,CONSUMER_SECRET are API keys given by twitter for developers

```python
username = "BarackObama"
```

Here "username" is which user data we want to get from the twitter

```python
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
```

Here tweepy checks authentication of us by our CONSUMER_KEY,CONSUMER_SECRET

```python
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
```

After authentication is done tweepy checks ACCESS_TOKEN and ACCESS_SECRET

```python
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True, compression=True)
```

Here tweepy.API() assigns to api this says that the user has authenticated,the next parameters controls tweepy to blocking and they waits untill next API calls retuns 

```python
with open('twitter_data.jsonl', 'w') as outfile: 
    for status in api.user_timeline(screen_name = username,count = 500):
        outfile.write(json.dumps(status._json)+"\n")
```

I had opened a json file named as midasIIITD_twitter_data.jsonl with write permission then I had used twitter timeline API to get tweets from the particular username and count defines number of tweets we want.Finally I'd dump the data I got from twitter to the file midasIIID_twitter_data.jsonl.
